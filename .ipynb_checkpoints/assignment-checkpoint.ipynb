## II. Sessa Empirical Estimator with K-means: Python Implementation

This section implements the SEE using K-Means to estimate prescription durations. We compute the time gap between consecutive prescriptions, build an ECDF (trimming the upper 20% to remove outliers), standardize these intervals, and then apply K-Means with the optimal cluster number chosen via Silhouette Analysis. The median interval of each cluster is used as the estimated duration.

**Algorithm Steps:**
- Compute event intervals between consecutive prescription fills.
- Construct the ECDF and trim the upper 20%.
- Randomly sample one prescription pair per patient.
- Standardize the intervals.
- Apply K-Means and select the optimal number of clusters using Silhouette Analysis.
- Estimate duration by computing the median interval per cluster.

```python
# =============================================================================
# Function: See
# =============================================================================
def See(arg1):
    # Filter dataset where ATC equals arg1
    C09CA01 = tidy[tidy['ATC'] == arg1].copy()
    
    # Take a random sequence of consecutive prescriptions in the dataset
    Drug_see_p0 = C09CA01.copy()
    Drug_see_p1 = C09CA01.copy()
    
    # Arrange by pnr and eksd, then compute the previous eksd per pnr
    Drug_see_p1 = Drug_see_p1.sort_values(by=['pnr', 'eksd'])
    Drug_see_p1['prev_eksd'] = Drug_see_p1.groupby('pnr')['eksd'].shift(1)
    
    # Remove rows with NA in prev_eksd
    Drug_see_p1 = Drug_see_p1[~Drug_see_p1['prev_eksd'].isna()].copy()
    
    # For each pnr, randomly sample one row without using groupby.apply
    groups = []
    for name, group in Drug_see_p1.groupby('pnr'):
        groups.append(group.sample(1, random_state=1234))
    Drug_see_p1 = pd.concat(groups).reset_index(drop=True)
    
    # Keep only the needed columns
    Drug_see_p1 = Drug_see_p1[['pnr', 'eksd', 'prev_eksd']].copy()
    
    # Compute event interval (in days)
    Drug_see_p1['event_interval'] = (Drug_see_p1['eksd'] - Drug_see_p1['prev_eksd']).dt.days.astype(float)
    
    # ---- Build ECDF using only positive event intervals ----
    valid_intervals = Drug_see_p1.loc[Drug_see_p1['event_interval'] > 0, 'event_interval']
    if len(valid_intervals) == 0:
        raise ValueError("No positive event intervals found.")
    sorted_vals = np.sort(valid_intervals.values)
    n = len(sorted_vals)
    ecdf_y = np.arange(1, n + 1) / n
    dfper = pd.DataFrame({'x': sorted_vals, 'y': ecdf_y})
    
    # Retain the lower 80% of the ECDF (i.e. where y <= 0.8)
    dfper_80 = dfper[dfper['y'] <= 0.8].copy()
    ni = dfper_80['x'].max() if not dfper_80.empty else dfper['x'].max()
    
    # Plot the 80% ECDF and the full (100%) ECDF side by side
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    axs[0].plot(dfper_80['x'], dfper_80['y'], marker='.', linestyle='none')
    axs[0].set_title(f"80% ECDF ({arg1})")
    axs[0].set_xlabel("Event Interval")
    axs[0].set_ylabel("ECDF")
    
    axs[1].plot(sorted_vals, ecdf_y, marker='.', linestyle='none')
    axs[1].set_title(f"100% ECDF ({arg1})")
    axs[1].set_xlabel("Event Interval")
    axs[1].set_ylabel("ECDF")
    plt.tight_layout()
    plt.show()
    
    # Plot frequency of 'pnr'
    m1 = Drug_see_p1['pnr'].value_counts()
    m1.plot(kind='bar', title=f"Frequency of pnr ({arg1})")
    plt.show()
    
    # --- Filter rows with event_interval <= ni ---
    Drug_see_p2 = Drug_see_p1[Drug_see_p1['event_interval'] <= ni].copy()
    
    # Filter out event_interval values that are zero or negative to avoid log(0) issues
    Drug_see_p2 = Drug_see_p2[Drug_see_p2['event_interval'] > 0].copy()
    
    # --- Density estimation of log(event_interval) ---
    log_vals = np.log(Drug_see_p2['event_interval'].values)
    density = gaussian_kde(log_vals)
    x_dens = np.linspace(log_vals.min(), log_vals.max(), 100)
    y_dens = density(x_dens)
    
    plt.figure()
    plt.plot(x_dens, y_dens)
    plt.title(f"Log(event interval) Density ({arg1})")
    plt.xlabel("Log(event interval)")
    plt.ylabel("Density")
    plt.show()
    
    # Create a DataFrame from the density estimation and scale it
    a_df = pd.DataFrame({'x': x_dens, 'y': y_dens})
    a_scaled = scale(a_df)  # returns a NumPy array
    
    # --- Silhouette Analysis to determine optimal number of clusters ---
    silhouette_scores = {}
    # Try number of clusters from 2 to min(10, number of samples)
    for k in range(2, min(10, len(a_scaled)) + 1):
        kmeans = KMeans(n_clusters=k, random_state=1234)
        labels = kmeans.fit_predict(a_scaled)
        score = silhouette_score(a_scaled, labels)
        silhouette_scores[k] = score
    if silhouette_scores:
        max_cluster = max(silhouette_scores, key=silhouette_scores.get)
    else:
        max_cluster = 1
    
    # Optionally plot silhouette scores
    plt.figure()
    plt.plot(list(silhouette_scores.keys()), list(silhouette_scores.values()), marker='o')
    plt.title(f"Silhouette Analysis ({arg1})")
    plt.xlabel("Number of clusters")
    plt.ylabel("Silhouette Score")
    plt.show()
    
    # --- K-means Clustering on dfper['x'] ---
    X_cluster = dfper['x'].values.reshape(-1, 1)
    kmeans_final = KMeans(n_clusters=max_cluster, random_state=1234)
    cluster_labels = kmeans_final.fit_predict(X_cluster)
    dfper['cluster'] = cluster_labels
    
    # Compute summary statistics (min, max, median) on log(x) by cluster
    dfper['log_x'] = np.log(dfper['x'])
    cluster_stats = dfper.groupby('cluster')['log_x'].agg(['min', 'max', 'median']).reset_index()
    # Exponentiate to revert from log scale
    cluster_stats['Minimum'] = np.exp(cluster_stats['min'])
    cluster_stats['Maximum'] = np.exp(cluster_stats['max'])
    cluster_stats['Median'] = np.exp(cluster_stats['median'])
    # Retain only clusters with Median > 0
    cluster_stats = cluster_stats[cluster_stats['Median'] > 0].copy()
    
    # --- Cross join Drug_see_p1 with cluster_stats ---
    Drug_see_p1['_key'] = 1
    cluster_stats['_key'] = 1
    results = pd.merge(Drug_see_p1, cluster_stats, on='_key').drop('_key', axis=1)
    # Assign Final_cluster where event_interval falls within the cluster's [Minimum, Maximum]
    results['Final_cluster'] = np.where(
        (results['event_interval'] >= results['Minimum']) & (results['event_interval'] <= results['Maximum']),
        results['cluster'], np.nan
    )
    results = results.dropna(subset=['Final_cluster']).copy()
    # Keep only the needed columns and rename for consistency
    results = results[['pnr', 'Median', 'cluster']].rename(columns={'cluster': 'Cluster'})
    
    # Determine the most frequent cluster from results
    cluster_freq = results['Cluster'].value_counts()
    if not cluster_freq.empty:
        most_freq_cluster = cluster_freq.idxmax()
        t1 = cluster_stats[cluster_stats['cluster'] == most_freq_cluster][['Median']].iloc[0]
    else:
        t1 = pd.Series({'Median': np.nan})
    
    # Merge the results back into Drug_see_p1 by 'pnr'
    Drug_see_p1 = pd.merge(Drug_see_p1, results, on='pnr', how='left')
    # Fill missing Median with t1's Median and missing Cluster with 0
    Drug_see_p1['Median'] = Drug_see_p1['Median'].fillna(t1['Median'])
    Drug_see_p1['Cluster'] = Drug_see_p1['Cluster'].fillna(0)
    # Calculate test as the rounded difference between event_interval and Median
    Drug_see_p1['test'] = np.round(Drug_see_p1['event_interval'] - Drug_see_p1['Median'], 1)
    
    # Extract the columns of interest
    Drug_see_p3 = Drug_see_p1[['pnr', 'event_interval', 'Median', 'Cluster']].copy()
    
    # --- Assign Duration ---
    Drug_see_p0 = pd.merge(Drug_see_p0, Drug_see_p3, on='pnr', how='left')
    Drug_see_p0['Median'] = Drug_see_p0['Median'].fillna(t1['Median'])
    Drug_see_p0['Cluster'] = Drug_see_p0['Cluster'].fillna(0)
    
    return Drug_see_p0
```

```python
# =============================================================================
# Function: see_assumption
# =============================================================================
def see_assumption(df, drug):
    # Arrange by pnr and eksd and compute the previous eksd
    df_sorted = df.sort_values(by=['pnr', 'eksd']).copy()
    df_sorted['prev_eksd'] = df_sorted.groupby('pnr')['eksd'].shift(1)
    
    # Create a sequential p_number per pnr
    df_sorted['p_number'] = df_sorted.groupby('pnr').cumcount() + 1
    # Keep only rows with p_number >= 2
    Drug_see2 = df_sorted[df_sorted['p_number'] >= 2].copy()
    Drug_see2 = Drug_see2[['pnr', 'eksd', 'prev_eksd', 'p_number']].copy()
    
    # Compute Duration in days
    Drug_see2['Duration'] = (Drug_see2['eksd'] - Drug_see2['prev_eksd']).dt.days.astype(float)
    # Convert p_number to a categorical variable (here, simply as string labels)
    Drug_see2['p_number'] = Drug_see2['p_number'].astype(str)
    
    # Plot a boxplot of Duration by p_number using seaborn
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='p_number', y='Duration', data=Drug_see2)
    plt.title(f"Boxplot of Duration by p_number ({drug})")
    
    # Calculate the median duration per pnr and then the overall median of these medians
    medians_of_medians = Drug_see2.groupby('pnr')['Duration'].median().median()
    plt.axhline(y=medians_of_medians, linestyle='--', color='red', label=f"Median: {medians_of_medians:.2f}")
    plt.legend()
    plt.show()
    
    return plt
```

```python
# =============================================================================
# Main Execution for SEE + Kmeans
# =============================================================================
result_kmeans_medA = See("medA")
result_kmeans_medB = See("medB")

see_assumption(result_kmeans_medA, "medA")
see_assumption(result_kmeans_medB, "medB")
```

## III. Sessa Empirical Estimator with DBSCAN: Alternative Clustering Algorithm

This section implements SEE using DBSCAN as an alternative to K-Means. The same pipeline is followed: compute event intervals, build and trim the ECDF, and standardize the data. DBSCAN is then applied to automatically detect clusters (and noise). The median interval of each cluster (excluding noise) is used as the estimated prescription duration.

**Algorithm Steps:**
- Compute event intervals between consecutive prescription fills.
- Build the ECDF and trim the upper 20%.
- Randomly sample one prescription pair per patient.
- Standardize the intervals.
- Apply DBSCAN (tuning parameters such as eps and min_samples).
- Estimate duration by computing the median interval per cluster (ignoring noise).

**The Evaluation of this alternative algorithm will be in the next section**

```python
# =============================================================================
# Function: See_dbscan
# =============================================================================
def See_dbscan(arg1, eps=0.15, min_samples=5):
    # Filter dataset where ATC equals arg1
    C09CA01 = tidy[tidy['ATC'] == arg1].copy()
    
    # Take a random sequence of consecutive prescriptions in the dataset
    Drug_see_p0 = C09CA01.copy()
    Drug_see_p1 = C09CA01.copy()
    
    # Arrange by pnr and eksd, then compute the previous eksd per pnr
    Drug_see_p1 = Drug_see_p1.sort_values(by=['pnr', 'eksd'])
    Drug_see_p1['prev_eksd'] = Drug_see_p1.groupby('pnr')['eksd'].shift(1)
    
    # Remove rows with NA in prev_eksd
    Drug_see_p1 = Drug_see_p1[~Drug_see_p1['prev_eksd'].isna()].copy()
    
    # For each pnr, randomly sample one row without using groupby.apply
    groups = []
    for name, group in Drug_see_p1.groupby('pnr'):
        groups.append(group.sample(1, random_state=1234))
    Drug_see_p1 = pd.concat(groups).reset_index(drop=True)
    
    # Keep only the needed columns
    Drug_see_p1 = Drug_see_p1[['pnr', 'eksd', 'prev_eksd']].copy()
    
    # Compute event interval (in days)
    Drug_see_p1['event_interval'] = (Drug_see_p1['eksd'] - Drug_see_p1['prev_eksd']).dt.days.astype(float)
    
    # ---- Build ECDF using only positive event intervals ----
    valid_intervals = Drug_see_p1.loc[Drug_see_p1['event_interval'] > 0, 'event_interval']
    if len(valid_intervals) == 0:
        raise ValueError("No positive event intervals found.")
    sorted_vals = np.sort(valid_intervals.values)
    n = len(sorted_vals)
    ecdf_y = np.arange(1, n + 1) / n
    dfper = pd.DataFrame({'x': sorted_vals, 'y': ecdf_y})
    
    # Retain the lower 80% of the ECDF (i.e. where y <= 0.8)
    dfper_80 = dfper[dfper['y'] <= 0.8].copy()
    ni = dfper_80['x'].max() if not dfper_80.empty else dfper['x'].max()
    
    # --- Filter rows with event_interval <= ni ---
    Drug_see_p2 = Drug_see_p1[Drug_see_p1['event_interval'] <= ni].copy()
    Drug_see_p2 = Drug_see_p2[Drug_see_p2['event_interval'] > 0].copy()  # Ensure positive values
    
    # --- Density estimation of log(event_interval) ---
    log_vals = np.log(Drug_see_p2['event_interval'].values)
    density = gaussian_kde(log_vals)
    x_dens = np.linspace(log_vals.min(), log_vals.max(), 100)
    y_dens = density(x_dens)
    
    # Create a DataFrame from the density estimation and scale it
    a_df = pd.DataFrame({'x': x_dens, 'y': y_dens})
    a_scaled = scale(a_df)  # returns a NumPy array
    
    # --- DBSCAN Clustering on dfper['x'] ---
    X_cluster = dfper['x'].values.reshape(-1, 1)
    db = DBSCAN(eps=eps, min_samples=min_samples)
    cluster_labels = db.fit_predict(X_cluster)
    dfper['cluster'] = cluster_labels
    
    # Exclude noise points (label = -1) for computing cluster summary statistics
    valid_clusters = dfper[dfper['cluster'] != -1].copy()
    valid_clusters['log_x'] = np.log(valid_clusters['x'])
    cluster_stats = valid_clusters.groupby('cluster')['log_x'].agg(['min', 'max', 'median']).reset_index()
    cluster_stats['Minimum'] = np.exp(cluster_stats['min'])
    cluster_stats['Maximum'] = np.exp(cluster_stats['max'])
    cluster_stats['Median'] = np.exp(cluster_stats['median'])
    cluster_stats = cluster_stats[cluster_stats['Median'] > 0].copy()
    
    # --- Cross join Drug_see_p1 with cluster_stats ---
    Drug_see_p1['_key'] = 1
    cluster_stats['_key'] = 1
    results = pd.merge(Drug_see_p1, cluster_stats, on='_key').drop('_key', axis=1)
    # Assign Final_cluster where event_interval falls within the cluster's [Minimum, Maximum]
    results['Final_cluster'] = np.where(
        (results['event_interval'] >= results['Minimum']) & (results['event_interval'] <= results['Maximum']),
        results['cluster'], np.nan
    )
    results = results.dropna(subset=['Final_cluster']).copy()
    results = results[['pnr', 'Median', 'cluster']].rename(columns={'cluster': 'Cluster'})
    
    # Determine the most frequent cluster from results
    cluster_freq = results['Cluster'].value_counts()
    if not cluster_freq.empty:
        most_freq_cluster = cluster_freq.idxmax()
        t1 = cluster_stats[cluster_stats['cluster'] == most_freq_cluster][['Median']].iloc[0]
    else:
        t1 = pd.Series({'Median': np.nan})
    
    # Merge the results back into Drug_see_p1 by 'pnr'
    Drug_see_p1 = pd.merge(Drug_see_p1, results, on='pnr', how='left')
    Drug_see_p1['Median'] = Drug_see_p1['Median'].fillna(t1['Median'])
    Drug_see_p1['Cluster'] = Drug_see_p1['Cluster'].fillna(0)
    Drug_see_p1['test'] = np.round(Drug_see_p1['event_interval'] - Drug_see_p1['Median'], 1)
    
    # Extract the columns of interest
    Drug_see_p3 = Drug_see_p1[['pnr', 'event_interval', 'Median', 'Cluster']].copy()
    
    # --- Assign Duration ---
    Drug_see_p0 = pd.merge(Drug_see_p0, Drug_see_p3, on='pnr', how='left')
    Drug_see_p0['Median'] = Drug_see_p0['Median'].fillna(t1['Median'])
    Drug_see_p0['Cluster'] = Drug_see_p0['Cluster'].fillna(0)
    
    return Drug_see_p0
```

## IV. Evaluations

In this section, we compare the clustering performance of the SEE implemented with K-Means versus DBSCAN using standard evaluation metrics. We report the number of clusters, noise percentage (for DBSCAN), Davies-Bouldin Index, Silhouette Score, and Calinski-Harabasz Index. These metrics help us assess cluster compactness, separation, and overall clustering quality, guiding us on which method better captures refill patterns in the data.

For the SEE + DBSCAN, the epsilon and min_samples are set with the following values:
- **medA -> eps = 3, min_samples = 2**
- **medB -> eps = 4, min_samples = 2**

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

def evaluate_clustering(kmeans_result, dbscan_result, drug):
    """
    Compare the performance of SEE (using KMeans) versus SEE_dbscan (using DBSCAN)
    using several evaluation metrics.

    Both DataFrames are expected to have:
      - "pnr": identifier
      - "event_interval": numeric event interval (in days)
      - "Cluster": cluster label (for DBSCAN, noise is assumed to be labeled as 0)
    """
    print(f"\nüîé **Clustering Evaluation for {drug}** üîé")
    
    # 1Ô∏è‚É£ Number of clusters (for DBSCAN, exclude noise (label 0))
    num_clusters_kmeans = kmeans_result["Cluster"].nunique()
    num_clusters_dbscan = dbscan_result[dbscan_result["Cluster"] != 0]["Cluster"].nunique()
    print(f"üîπ K-Means Clusters: {num_clusters_kmeans}")
    print(f"üîπ DBSCAN Clusters (excluding noise): {num_clusters_dbscan}")
    
    # 2Ô∏è‚É£ DBSCAN Noise Percentage (noise labeled as 0)
    noise_count = (dbscan_result["Cluster"] == 0).sum()
    total_count = len(dbscan_result)
    noise_percentage = (noise_count / total_count) * 100
    print(f"üîπ DBSCAN Noise Percentage: {noise_percentage:.2f}%")
    
    # 3Ô∏è‚É£ Davies-Bouldin Index (lower is better)
    kmeans_filtered = kmeans_result.dropna(subset=["event_interval", "Cluster"])
    dbscan_filtered = dbscan_result[dbscan_result["Cluster"] != 0].dropna(subset=["event_interval", "Cluster"])
    
    if len(kmeans_filtered) > 1 and len(dbscan_filtered) > 1:
        db_kmeans = davies_bouldin_score(kmeans_filtered[["event_interval"]], kmeans_filtered["Cluster"])
        db_dbscan = davies_bouldin_score(dbscan_filtered[["event_interval"]], dbscan_filtered["Cluster"])
        print(f"üîπ Davies-Bouldin Index (K-Means): {db_kmeans:.2f}")
        print(f"üîπ Davies-Bouldin Index (DBSCAN): {db_dbscan:.2f}")
    else:
        print("‚ùå Davies-Bouldin Score cannot be computed (insufficient valid clusters).")
    
    # 4Ô∏è‚É£ Silhouette Score (higher is better)
    if len(kmeans_filtered) > 1 and len(dbscan_filtered) > 1:
        sil_kmeans = silhouette_score(kmeans_filtered[["event_interval"]], kmeans_filtered["Cluster"])
        sil_dbscan = silhouette_score(dbscan_filtered[["event_interval"]], dbscan_filtered["Cluster"])
        print(f"üîπ Silhouette Score (K-Means): {sil_kmeans:.2f}")
        print(f"üîπ Silhouette Score (DBSCAN): {sil_dbscan:.2f}")
    else:
        print("‚ùå Silhouette Score cannot be computed (insufficient valid clusters).")
    
    # 5Ô∏è‚É£ Calinski-Harabasz Index (higher is better)
    if len(kmeans_filtered) > 1 and len(dbscan_filtered) > 1:
        ch_kmeans = calinski_harabasz_score(kmeans_filtered[["event_interval"]], kmeans_filtered["Cluster"])
        ch_dbscan = calinski_harabasz_score(dbscan_filtered[["event_interval"]], dbscan_filtered["Cluster"])
        print(f"üîπ Calinski-Harabasz Index (K-Means): {ch_kmeans:.2f}")
        print(f"üîπ Calinski-Harabasz Index (DBSCAN): {ch_dbscan:.2f}")
    else:
        print("‚ùå Calinski-Harabasz Score cannot be computed (insufficient valid clusters).")
```

```python
# Evaluate and compare clustering performance
result_dbscan_medA = See_dbscan("medA", 3, 2)
result_dbscan_medB = See_dbscan("medB", 4, 2)

evaluate_clustering(result_kmeans_medA, result_dbscan_medA, "medA")
evaluate_clustering(result_kmeans_medB, result_dbscan_medB, "medB")
```

## V. Insights

This section summarizes the insights drawn from the clustering evaluations. We interpret the evaluation metrics to understand how each method differentiates refill behaviors. For instance, while K-Means assigns all patients to clusters, DBSCAN may label a significant portion as noise, highlighting dense versus sparse refill patterns. We also explore how the estimated prescription durations relate to patient adherence and medication use trends, providing actionable insights into treatment patterns.

For medA and medB the results show different trade‚Äêoffs between the two methods:

**Observations and Insight for medA**
- **K-Means:** Partitions medA into 2 clusters with moderate separation (Silhouette‚ÄØ=‚ÄØ0.64, DB Index‚ÄØ=‚ÄØ0.54, CH‚ÄØ=‚ÄØ1033).
- **DBSCAN:** Extracts 16 clusters with 38.07% noise and excellent internal quality (Silhouette‚ÄØ=‚ÄØ0.80, DB Index‚ÄØ=‚ÄØ0.17, CH‚ÄØ=‚ÄØ15661).
- **Insight:** DBSCAN isolates highly dense, well-separated subgroups from medA at the cost of excluding a significant portion of the data, which is useful if the focus is on core refill patterns rather than full data coverage.

**Observations and Insight for medB**
- **K-Means:** Clusters medB into 2 groups with strong overall separation (Silhouette‚ÄØ=‚ÄØ0.76, DB Index‚ÄØ=‚ÄØ0.49, CH‚ÄØ=‚ÄØ654).
- **DBSCAN:** Identifies 6 clusters with lower noise (17.89%) and tighter clusters (DB Index‚ÄØ=‚ÄØ0.23) but with slightly reduced global separation (Silhouette‚ÄØ=‚ÄØ0.59, CH‚ÄØ=‚ÄØ286).
- **Insight:** For medB, K-Means provides a more cohesive overall partition, while DBSCAN's finer subdivision may capture additional nuances but risks over-segmentation.

## VI. Conclusion

In conclusion, our analysis of the Sessa Empirical Estimator applied to medA and medB reveals distinct trade-offs between clustering algorithms: K-Means consistently partitions the entire dataset into 2 clusters with robust global separation, while DBSCAN, although producing finer subclusters (16 clusters for medA and 6 for medB) and exhibiting superior internal metrics (e.g., higher Silhouette scores and lower Davies-Bouldin Indices), tends to exclude a substantial portion of the data as noise, particularly for medA, where 38% of points are classified as noise. **These results indicate that if the goal is to capture overall structure and ensure full data coverage, K-Means may be preferable;** however, if the **focus is on isolating highly dense, homogeneous core groups, DBSCAN offers enhanced cluster quality** despite higher noise levels.
